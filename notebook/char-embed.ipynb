```
def read_char_vocab(file_path):
  char_vocab = []
  for line in open(file_path, encoding='utf-8'):
    char_vocab.append(line.splitlines()[0].lower())      
  return char_vocab
```

```
def char_to_int(char_vocab_data):
  dic = {}
  index = -1
  for char in char_vocab_data:
    try:
      dic[char]
    except:
      index = index + 1
      dic[char]=index 
  return dic
def int_to_char(char_vocab_data):
  dic = {}
  index = -1
  for char in char_vocab_data:
    try:
      dic[char]
    except:
      index = index + 1
      dic[index]=char
  return dic
char_to_int = char_to_int(char_vocab_data)
int_to_char = int_to_char(char_vocab_data)
```

```
import csv
import numpy

# đọc dữ liệu từ file csv sau đó chia dữ liệu làm tập input và output

input_train = []
output_train = []
with open('/content/drive/MyDrive/data/data_train_test.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        row = row[0].split(",")
        input_train.append(row[0])
        output_train.append(row[1])
        
print(input_train)
print(output_train)
```

```
# convert về dạng array
data_input = numpy.array(input_train)
data_output = numpy.array(output_train)


data_input_encode = []
data_output_encode = []
# print(data_input)

def convert_word_to_int(data):
  data_encode = []
  for word in data:
    temp = []
    for char in word:
      temp.append(char_to_int[char])
    data_encode.append(temp)
  return data_encode

data_input_encode = convert_word_to_int(data_input)
data_output_encode = convert_word_to_int(data_output)
print(data_input_encode)
print(data_output_encode)

# data_input_encode = numpy.array(data_input_encode)
# print(data_input_encode)
    
```

```
import string

# chuyển những từ đã được convert sang số  vào vector từ 10 chiều (tương ứng với 10 ký tự trên 1 từ)

data_array_input = []
data_array_output = []

def assign_data_to_array(data):
  data_array = []
  for i in data:
    char_emb = numpy.zeros((10,), dtype=int)
    for j in range(len(i)):
      char_emb[j] = i[j]
    data_array.append(char_emb)
  return data_array

data_array_input = assign_data_to_array(data_input_encode)
data_array_output = assign_data_to_array(data_output_encode)
print(data_array_input)
print(data_array_output)
```

```
data_train_input = []
data_train_output = []

# convert từ mảng vector từ sang data có thể  train được

def convert_to_data_train(data):
  data_train = []
  for i in data:
    char_emb = numpy.zeros((10, 137), dtype=int)
    for j in range(len(i)):
      char_emb[j][i[j]] = 1
    data_train.append(char_emb)
  return data_train

data_train_input = convert_to_data_train(data_array_input)
data_train_output = convert_to_data_train(data_array_output)

# convert từ object sang array
data_train_input = numpy.array(data_train_input)
data_train_output = numpy.array(data_train_output)

print(data_train_input.shape)
print(data_train_output.shape)
```

```
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, Masking
from keras.callbacks import ModelCheckpoint
from keras.utils import np_utils
import os
```

```
from keras.models import Sequential, Model, Input
from keras.layers import LSTM, Dense, TimeDistributed, Activation, Bidirectional, Masking, Embedding, Dropout, Flatten, concatenate, Conv1D, MaxPool1D
from keras import backend as K
from keras.utils import plot_model
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```
model = Sequential()
model.add(Masking(mask_value=0., input_shape=(10, 137)))
model.add(Bidirectional(LSTM(units=32, return_sequences=True, dropout=0.2,recurrent_dropout=0.2), merge_mode='concat'))
model.add(TimeDistributed(Dense(137)))
model.add(Activation('softmax'))
plot_model(model, "demo.png", show_shapes=True)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```
model.fit(data_train_input, 
          data_train_output,  
          epochs=50,
          batch_size = 16)
```
